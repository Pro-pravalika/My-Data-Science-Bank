### Plots in a Data science... 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

 ####  1.Barplot...

```python
plt.figure(figsize=(10,12))
ax = sns.countplot('day',data = df)
ax
```

### 2.Scatter plot...

```python
fig = plt.figure(figsize = (10, 10))
axes_1 = fig.add_subplot(1,1,1)
axes_1.scatter(x = df['total_bill'], y = df['tip'],color='black')
axes_1.set_facecolor('gray')

axes_1.set_title("Total Bill Vs Tip Colored by Sex and sized by size",size=30,color='black')
axes_1.set_xlabel("Total Bill",color='blue',size = '30')
axes_1.set_ylabel("Tip",color='blue',size = '30')
```

### 3. Reg plot

```python
plt.figure(figsize = (10,8))
sns.regplot(x = "total_bill", y = "tip", data = df,color='brown')
```

### 4.lmplot

```python
plt.figure(figsize = (10,8))
sns.lmplot(x = "total_bill", y = "tip", data = df)
```

### 5. Joinplot..

```python
plt.figure(figsize = (10,8))
sns.jointplot(x = "total_bill", y = "tip", data = df, kind = "hex",color='black')
plt.suptitle("Seaborn visualition",size=15,color='black')
```

```python
plt.figure(figsize = (10,8))

sns.kdeplot(data = df['total_bill'], data2 = df['tip'], shade = True,color='darkorange')
```

### 6.Boxplot.

###  Boxplot : boxplot is statistical visualisation of data. Boxplot show the outlier . 

```python
plt.figure(figsize = (10,8))
sns.boxplot(x = "time", y = "total_bill", data = df)
```

### 7. Violin plot .....

```python
plt.figure(figsize = (10,8))
sns.violinplot(x = "time", y = "total_bill", data = df)
```

### 8. Pair plot.

```python
sns.pairplot(df)  # Create the default pairplot.
```

```python
pair_grid = sns.PairGrid(df)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug = True)
```

### 9. Heatmap

```python
plt.figure(figsize = (10,8))
sns.heatmap(df.corr(), annot=True, cmap="tab20")
```

```python
plt.figure(figsize = (10,8))
facet = sns.FacetGrid(df, col = "day", hue = "sex")
facet.map(plt.scatter, "total_bill", "tip")

facet.add_legend()
```

```python
plt.figure(figsize = (10,8))
facet = sns.FacetGrid(df, col = "time", hue = "smoker")
facet.map(plt.scatter, "total_bill", "tip")
facet.add_legend()
```

```python
plt.figure(figsize = (10,8))
facet = sns.FacetGrid(df, col = "sex", hue = "day")
facet.map(plt.scatter, "total_bill", "tip")
facet.add_legend()
```
